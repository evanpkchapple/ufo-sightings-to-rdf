```python
import rdflib
import pyshacl


p1 = rdflib.Namespace("https://idmc.univ-lorraine.fr/sw/p1#")
rdf = rdflib.Namespace("http://www.w3.org/1999/02/22-rdf-syntax-ns#")
rdfs = rdflib.Namespace("http://www.w3.org/2000/01/rdf-schema#")
owl = rdflib.Namespace("http://www.w3.org/2002/07/owl#")
xsd = rdflib.Namespace("http://www.w3.org/2001/XMLSchema#")
sh = rdflib.Namespace("http://www.w3.org/ns/shacl#")


shacl_graph = rdflib.Graph()
shacl_graph.bind("p1", p1)
shacl_graph.bind("rdf", rdf)
shacl_graph.bind("rdfs", rdfs)
shacl_graph.bind("owl", owl)
shacl_graph.bind("xsd", xsd)
shacl_graph.bind("sh", sh)

# SHACL shape for p1:time
shacl_graph.add((p1.p1TimeShape, rdf.type, sh.NodeShape))
shacl_graph.add((p1.p1TimeShape, sh.targetClass, p1.Event))

# Property shape for p1:time
property_shape = rdflib.BNode()
shacl_graph.add((p1.p1TimeShape, sh.property, property_shape))
shacl_graph.add((property_shape, sh.path, p1.time))
shacl_graph.add((property_shape, sh.datatype, xsd.time))
shacl_graph.add((property_shape, sh.maxCount, rdflib.Literal(1)))


data_graph = rdflib.Graph()
data_graph.parse("/content/project.rdf", format="turtle")


conforms, report_graph, _ = pyshacl.validate(data_graph, shacl_graph=shacl_graph, inference="rdfs", debug=True)

if conforms:
    print("Data graph conforms to the SHACL shape constraints.")
else:
    print("Data graph does not conform to the SHACL shape constraint.")

print("\nValidation Report:")
print(report_graph)
```
